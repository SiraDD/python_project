```python
df['artist'] = df['artist'].astype(str)
df['song'] = df['song'].astype(str)
df['genre'] = df['genre'].astype(str)
```

```python
df.dtypes
```

Fixing problem with genre

```python
#Clean data
df['genre'].unique()
```

```python
dct = {}
for i in range(0, len(df)):
  lst = df.loc[i]['genre'].split(',')
  lst = [i.strip() for i in lst]
  dct[df.loc[i]['genre']] = lst;

df['genre_list'] = df['genre'].map(dct)
df = df.drop(columns=["genre"])
df
```

```python
for year in df['year'].unique():
  print(year, df.loc[df['year']==year]['song'].count())
```

```python
# Only count years that contain atleast 5 data values
# Remove 1998 and 2020
df = df.drop(df[df['year']==1998].index)
df = df.drop(df[df['year']==2020].index)
```

```python
df_per_year = df.groupby(['year']).mean()
```

Regression 
Some meaningful correlation: 
- Explicit and energy (-0.73)
- Acousticness and energy (-0.76)
- Loudness and energy (0.67)
- Danceability and energy (-0.64)
- liveness and energy (0.60)
- Popularity and acousticness (0.37)
- Valence and duration (0.66)

```python
plt.scatter(df_per_year['explicit'], df_per_year['energy'])
```

```python

params = np.polyfit(df_per_year["explicit"],df_per_year["energy"],deg=1)
```

```python
xs = np.linspace(df_per_year["explicit"].min(),df_per_year["energy"].max(), 1000)
ys = params[0]*xs + params[1]

plt.scatter(df_per_year['explicit'], df_per_year['energy'])
plt.plot(xs, ys, color="red")
```
